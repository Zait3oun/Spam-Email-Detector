```python
data.drop(columns=['Unnamed: 0','label'],axis=1)
```

```python
X = data['text']
y = data['label_num']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vecmodel=TfidfVectorizer(stop_words='english',
    lowercase=True,
    ngram_range=(1,2)
)
X_vec = vecmodel.fit_transform(X)
X_vec = pd.DataFrame.sparse.from_spmatrix(X_vec)# عشان تتعمل داتا فريم 
```

```python
print(f"the new shape for x is {X_vec.shape}")
X_vec.head()
```

```python
for i in range(5):
    print(sorted(list(X_vec.iloc[i,:]),reverse=True)[:20])
    print('--------------')
```

```python
x_train,x_test,y_train,y_test=train_test_split(X_vec,y,test_size=0.25,random_state=42)
```

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_test,y_pred)
accuracy
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, y_pred)
cm
```

```python
from sklearn.metrics import classification_report

print(classification_report(y_test, y_pred))
```

```python
from sklearn.linear_model import LogisticRegression
```
